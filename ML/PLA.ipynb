Testing - 25% data <br/>

```python
import numpy as np
import pandas as pd

# Load your data from CSV
data = pd.read_csv('data1.csv')

X = data[['x1', 'x2']]
y = data['Class']

# Shuffle the indices
indices = np.arange(len(data))
np.random.shuffle(indices)

# Calculate the split index for 75% training data
split_index = int(0.75 * len(data))

# Use the shuffled indices to create training and testing sets
X_train = X.iloc[indices[:split_index]].to_numpy()
y_train = y.iloc[indices[:split_index]].to_numpy()
X_test = X.iloc[indices[split_index:]].to_numpy()
y_test = y.iloc[indices[split_index:]].to_numpy()

class Perceptron:
    def __init__(self, learning_rate, epochs):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.updates = 0

    def train(self, X, y, X_test, y_test):
        X = np.c_[np.ones(X.shape[0]), X]
        X_test = np.c_[np.ones(X_test.shape[0]), X_test]
        self.weights = np.zeros(X.shape[1])

        for _ in range(self.epochs):
            for xi, target in zip(X, y):
                prediction = self.predict(xi)
                if prediction != target:
                    update = self.learning_rate * (target - prediction)
                    self.weights += update * xi
                    self.updates += 1

                    # Calculate accuracy on training set at each stage
                    y_train_pred = self.predict(X)
                    accuracy_train = np.mean(y == y_train_pred)

                    # Calculate accuracy on test set at each stage
                    y_test_pred = self.predict(X_test)
                    accuracy_test = np.mean(y_test == y_test_pred)

                    # Print accuracy at each stage
                    print(f"Update {self.updates} - Accuracy on training set: {accuracy_train}, Accuracy on test set: {accuracy_test}")

        # Print final weights and number of updates
        print("Final Weights [bias : w(x1) : w(x2)]: \n\t", self.weights)
        print("No. of updates:", self.updates)

    def predict(self, X):
        return np.where(np.dot(X, self.weights) >= 0, 1, -1)

# Train the perceptron with a learning rate of 1
perceptron = Perceptron(learning_rate=0.1, epochs=1000)
perceptron.train(X_train, y_train, X_test, y_test)
```
